```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
siec = torch.nn.Sequential(
    torch.nn.Linear(1, 5),
    torch.nn.Sigmoid(),
    torch.nn.Linear(5, 5),
    torch.nn.Sigmoid(),
    torch.nn.Linear(5,1),
)
optim = torch.optim.Adam(params=siec.parameters(), lr = 0.01)
df = pd.read_csv("/home/pawel/studia/MIOwAD/MIOwAD_labs/mio1/regression/square-simple-test.csv")
X = torch.from_numpy(np.asarray(df["x"]).reshape((-1,1))).float()
y = torch.from_numpy(np.asarray(df["y"]).reshape((-1,1))).float()
for step in range(1,100000):
    y_hat = siec(X)
    loss = (y_hat-y).pow(2).mean()
    print(step, loss.item())
    optim.zero_grad()
    loss.backward()
    # optim.step(lambda : (siec(X)-y).pow(2).mean())
    optim.step()


```
